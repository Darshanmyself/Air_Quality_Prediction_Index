**Handeling the missing values**

Converting all -200 to NaN value.


Replace all Nan values with the mean of that specific column.

```python
air_quality_data=air_quality_data.replace(to_replace=-200,value=np.NaN)
```

```python
air_quality_data.isnull().sum()
```

This shows the actual number of missing values.

```python
air_quality_data.tail(5)
```

```python
air_quality_data.iloc[:,2:].mean()
```

```python
#Replacing the missing value with mean value
air_quality_data=air_quality_data.fillna(air_quality_data.iloc[:,2:].mean())
```

```python
air_quality_data.isnull().sum()
```

**Forecasting with FB Prophet Algorithm**

```python
# Converting the date from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_data['Date'],format='mixed')
print(date_info)
```

```python
time_info=air_quality_data['Time'].apply(lambda x:x.replace('.',':'))
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# Combining 2 series to a pandas dataframe
date_time=pd.concat([date_info,time_info],axis=1)
date_time.head(5)
```

```python
#combining date and time
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
date_time
```

```python
date_time.info()
```

We have to convert ds from object datatype to date time format

```python
date_time['ds']=pd.to_datetime(date_time['ds'])
date_time.info()
```

```python
data=pd.DataFrame()
```

```python
data['ds']=date_time['ds']
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
pip install prophet
```

```python
from prophet import Prophet
```

```python
# training the Prophet Model
model=Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
plot_plotly(model, forecast)
```

```python
plot_components_plotly(model, forecast)
```
